```python
import pandas as pd
```

```python
path_eyes='output/eyes'
eyes_classification = pd.read_csv(path_eyes+'/eyes_classification.csv', index_col=0)
```

```python
eyes_classification
```

```python
def eye_classifier(eye_open, threshold=0.0015):
    if eye_open > 1 - threshold:      
        return 'open'
    elif eye_open < threshold:
        return 'closed'
    else:
        return 'unknown'
```

```python
eyes_classification['classification'] = eyes_classification.open.apply(lambda x: eye_classifier(x))
```

```python
eyes_classification
```

```python
eyes_classification[eyes_classification.classification=='unknown'].shape[0]
```

```python
picture_names = eyes_classification.name.unique()
picture_names
```

```python
def face_classifier(eye_values):
    eye1 = eye_values[0]
    eye2 = eye_values[1]
    if eye1 == 'open' and eye2 == 'open':
        return 'open'
    if eye1 == 'closed' and eye2 == 'closed':
        return 'closed'
    else:
        return 'unknown'
```

```python
df = pd.DataFrame()
picture_names = eyes_classification.name.unique()
for picture_name in picture_names:
    df_picture = eyes_classification[eyes_classification.name == picture_name]
    faces = df_picture.num.unique()
    for face_num in faces:
        df_faces = df_picture[df_picture.num == face_num]
        classification = face_classifier(df_faces.classification.values)
        df = pd.concat([df, pd.DataFrame({'name':picture_name, 'face':face_num, 'classification': classification},[0])])
```
