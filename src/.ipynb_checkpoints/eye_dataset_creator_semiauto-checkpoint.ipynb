```python
def get_eyes(face_image, amplitude=0.45, heihgt=0.1, wide=0.1):
    eye_right = face_image[round(face_image.shape[0]*heihgt):round(face_image.shape[0]*(heihgt+amplitude)), 
                      round(face_image.shape[1]*wide):round(face_image.shape[1]*(amplitude+wide))]
    
    eye_left = face_image[round(face_image.shape[0]*heihgt):round(face_image.shape[0]*(heihgt+amplitude)), 
                      round(face_image.shape[1]*(1-(amplitude+wide))):round(face_image.shape[1]*(1-wide))]
    return eye_right, eye_left
```

```python
def create_folder(directory):
    path = directory
    # Check whether the specified path exists or not
    isExist = os.path.exists(path)
    if not isExist: 
        # Create a new directory because it does not exist 
        os.makedirs(path)
        print("The new directory is created!")
```

# When you have a folder with faces that you have to label

```python
def create_folders(directory_save_eyes):
    create_folder(directory_save_eyes+'/open')
    create_folder(directory_save_eyes+'/closed')
```

```python
def label_transform(label):
    if label == '00':
        right_lab = 'closed' 
        left_lab = 'closed'

    elif label == '01':
        right_lab = 'closed' 
        left_lab = 'open'

    elif label == '10':
        right_lab = 'open' 
        left_lab = 'closed' 

    elif label == '11':
        right_lab = 'open' 
        left_lab = 'open'
    
    else:
        right_lab = False
        left_lab = False
        
    return right_lab, left_lab
```

```python
def label_eyes(image, image_name, directory_save_eyes):
    eye_right, eye_left  = get_eyes(image)
    eye_mirror = cv2.flip(eye_left, 1)
    
    
    cv2.namedWindow('Image', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('Image', 600,600)
    cv2.imshow("Image", image)
    
    key = cv2.waitKey(100)#pauses for 3 seconds before fetching next image

    if key == 27:#if ESC is pressed, exit loop
        cv2.destroyAll

    label = input("Insert Closed or Open eyes (00-01-10-11) :")

    cv2.destroyAllWindows()

    right_lab, left_lab = label_transform(label)
    
    if right_lab and left_lab:
        cv2.imwrite(directory_save_eyes+'/'+right_lab+'/eye_right_'+image_name, eye_right)                    
        cv2.imwrite(directory_save_eyes+'/'+left_lab+'/eye_mirror_'+image_name, eye_mirror)
```

```python
def label_eyes_from_directory(directory_faces, directory_save_eyes, label='no_label'):

    create_folders(directory_save_eyes)
    
    image_list = os.listdir(directory_faces)
    
    for image_name in image_list:
        
        im_bgr = cv2.imread(directory_faces+'/'+image_name)
        image = im_bgr[:, :, ::-1]
        
        label_eyes(image, image_name, directory_save_eyes)
```

# Get faces from a bigger picture and label them

```python
import matplotlib.pyplot as plt
import face_recognition
import numpy as np
import tensorflow as tf
import os
```

```python
def label_group_picture(directory, image_name, directory_save_eyes):
    imagePath = directory + image_name
    
    image = face_recognition.load_image_file(imagePath)
    
    face_locations = face_recognition.face_locations(image)

    print("I found {} face(s) in this photograph.".format(len(face_locations)))

    num = 0

    create_folders(directory_save_eyes)

    for face_location in face_locations:

        # Print the location of each face in this image
        top, right, bottom, left = face_location

        # You can access the actual face itself like this:
        im_bgr = image[top:bottom, left:right]
        face_image = im_bgr[:, :, ::-1]

        label_eyes(face_image, image_name+'_'+str(num), directory_save_eyes)

        num += 1
```

```python
def label_eyes_from_dir_group(directory_pictures, directory_save_eyes):
    
    create_folders(directory_save_eyes)

    image_list = os.listdir(directory_pictures)
    
    for image_name in image_list:
        
        if image_name != 'done':      
            print('=================================')
            print(image_name)

            label_group_picture(directory_pictures, image_name, directory_save_eyes)

            shutil.move(directory_pictures+image_name, directory_pictures+'done/'+image_name)
    
```

### Label only the unknown pictures

```python
def label_transform_auto(eye_right_open, eye_mirror_open):
    if eye_right_open > 0.5:
        right_lab = 'open' 
    else:
        right_lab = 'closed'

    if eye_mirror_open > 0.5:
        mirror_lab = 'open' 
    else:
        mirror_lab = 'closed'
        
    return right_lab, mirror_lab
```

```python
def eye_unknown(eye_open, threshold):
    unknown = True
    if eye_open > (1-threshold) or eye_open < threshold:
        unknown = False
        
    return unknown
```

```python
def store_images(directory_save_eyes, image_name, eye_right, eye_mirror, right_lab, mirror_lab):
    
    # Store it in rgb
    eye_right = eye_right[:, :, ::-1]
    eye_mirror = eye_mirror[:, :, ::-1]
    
    cv2.imwrite(directory_save_eyes+'/'+right_lab+'/eye_right_'+image_name, eye_right)                    
    cv2.imwrite(directory_save_eyes+'/'+mirror_lab+'/eye_mirror_'+image_name, eye_mirror)
```

```python
def label_eyes_auto(image, image_name, directory_save_eyes, eye_right_open, eye_mirror_open, show=False):

    right_lab, mirror_lab = label_transform_auto(eye_right_open, eye_mirror_open)
    
    eye_right, eye_left  = get_eyes(image)
    eye_mirror = cv2.flip(eye_left, 1)
      
    store_images(directory_save_eyes+'_auto', image_name, eye_right, eye_mirror, right_lab, mirror_lab)
    
```

```python
def label_eyes_manual(image, image_name, directory_save_eyes):
    cv2.namedWindow('Image', cv2.WINDOW_NORMAL)
    cv2.resizeWindow('Image', 600,600)
    cv2.imshow("Image", image[:, :, ::-1])
    
    key = cv2.waitKey(100)#pauses for 3 seconds before fetching next image

    if key == 27:#if ESC is pressed, exit loop
        cv2.destroyAll

    label = input("Insert Closed or Open eyes (00-01-10-11) :")

    cv2.destroyAllWindows()

    right_lab, mirror_lab = label_transform(label)
    
    if right_lab and mirror_lab:
        eye_right, eye_left  = get_eyes(image)
        eye_mirror = cv2.flip(eye_left, 1)
        store_images(directory_save_eyes+'_manual', image_name, eye_right, eye_mirror, right_lab, mirror_lab)
```

```python
def label_eyes_semiauto(image, image_name, directory_save_eyes, model_sunglasses, model_right, threshold=0.1, show=False):
    eye_right, eye_left  = get_eyes(image)
    eye_mirror = cv2.flip(eye_left, 1)
    
    image1 = tf.image.resize(image, (160, 160))
    sunglasses = lib.model_lib_sunglasses.uses_sunglasses(image1, model_sunglasses)
    
    image2 = tf.image.resize(image, (320, 320))
    eye_right_open, eye_mirror_open = lib.model_lib_eye.is_open_face(image2, model_right, show=show)
    
    if not sunglasses:  
        if eye_unknown(eye_right_open, threshold) or eye_unknown(eye_mirror_open, threshold):   
            print('=================================================================')
            print('Manually sort eyes: ', eye_right_open, eye_mirror_open)
            label_eyes_manual(image, image_name, directory_save_eyes)

        else:
            print('=================================================================')
            print('Automatically stored eyes with: ', eye_right_open, eye_mirror_open)
            label_eyes_auto(image, image_name, directory_save_eyes, eye_right_open, eye_mirror_open)

```

```python
def label_group_picture_semiauto(directory, image_name, directory_save_eyes, threshold=0.1):
    imagePath = directory + image_name
    
    image = face_recognition.load_image_file(imagePath)
    
    face_locations = face_recognition.face_locations(image)

    print("I found {} face(s) in this photograph.".format(len(face_locations)))

    num = 0

    create_folders(directory_save_eyes+'_auto')
    create_folders(directory_save_eyes+'_manual')
    
    model_sunglasses = lib.model_lib.load_model('model/model_sunglasses')
    model_right = lib.model_lib.load_model('model/model_eye_right')

    for face_location in face_locations:

        # Print the location of each face in this image
        top, right, bottom, left = face_location

        # You can access the actual face itself like this:
        face_image = image[top:bottom, left:right]

        label_eyes_semiauto(face_image, 
                            image_name+'_'+str(num), 
                            directory_save_eyes, 
                            model_sunglasses, 
                            model_right, 
                            threshold=threshold)

        num += 1
```

```python
def label_eyes_from_dir_group(directory_pictures, directory_save_eyes, threshold=0.1):
    
    create_folders(directory_save_eyes)

    image_list = os.listdir(directory_pictures)
    
    for image_name in image_list:
        
        if image_name != 'done':      
            print('=================================')
            print(image_name)

            label_group_picture_semiauto(directory_pictures, image_name, directory_save_eyes, threshold=threshold)

            shutil.move(directory_pictures+image_name, directory_pictures+'done/'+image_name)
```
